```python
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from tqdm import tqdm
```

```python
### Indlæser data ###
PRL = pd.read_csv("/content/drive/MyDrive/Speciale_2024/1_input/preprocessed_PRL.csv", encoding='latin-1') # Specify the correct encoding
```

```python
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Your dictionary of keywords (lowercased)
keywords = [
    "palestinians", "palestinian", "palestine", "israel", "israeli", "israelis"
    "gaza", "west bank", "two-state solution", "hostage",
    "hostages", "hamas", "hama", "netanyahu"
]

# Function to lemmatize and check for keywords
def contains_keywords(text, keywords):
    # Tokenize and lemmatize the text
    words = text.split()  # Split text into words
    lemmatized_words = [lemmatizer.lemmatize(word.lower()) for word in words]
    # Check if any keyword is present
    return int(any(word in keywords for word in lemmatized_words))

# Ensure the 'text' column is of string type
PRL['text'] = PRL['text'].astype(str)

# Process the dataframe
tqdm.pandas(desc="Processing")  # Initialize progress bar description

# Lowercase and check for conflict-related statements
PRL['conflict_related'] = PRL['text'].progress_apply(lambda x: contains_keywords(x, keywords))

# Display the updated dataframe
PRL.head(10)
```

```python
# Count the number of conflict-related statements
conflict_related_count = PRL['conflict_related'].sum()

# Print the count
print(f"Number of conflict-related statements: {conflict_related_count}")
```
